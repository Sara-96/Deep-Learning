```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.random.randn(10000) + 10

# the histogram of the data
n, bins, patches = plt.hist(x1, 50, density=True, facecolor='b', alpha=0.5)

plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of:}\ \mu=10,\ \sigma=1$')
plt.grid(True)
plt.show()
```

```python
x1Sample = np.random.choice(x1,200)
n, bins, patches = plt.hist(x1Sample, 50, density=True, facecolor='r', alpha=0.5)

plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ Sample:}\ \mu=10,\ \sigma=1$')
plt.grid(True)
plt.show()
```

```python
x2 = 2 * np.random.rand(10000) - 1

# the histogram of the data
n, bins, patches = plt.hist(x2, 50, density=True, facecolor='b', alpha=0.5)

plt.ylabel('Probability')
plt.title('Histogram of Uniform data')
plt.grid(True)
plt.show()
```

```python
x3 = np.multiply(x1,x2)

# the histogram of the data
n, bins, patches = plt.hist(x3, 50, density=True, facecolor='b', alpha=0.5)

plt.ylabel('Probability')
plt.title('Histogram')
plt.grid(True)
plt.show()
```
