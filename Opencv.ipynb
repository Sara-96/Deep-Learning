```python
import cv2 
  
# Using cv2.imread() method 
img2 = cv2.imread('img.jpg') 
  
# Displaying the image
cv2.imshow('image',img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
a = np.array(img2)

print(a.shape)   # tuple, array shape
x, y, z = a.shape
print(a.ndim)    # int, number of array dimension
print(a.dtype)   # Data-type of the array’s elements
print(a.size)    # Number of elements in the array
```

```python
xx = np.arange(11,x-10)
yy = np.arange(11,y-10)

img_filtered = a

for i in xx:
    for j in yy:
        H0 = a[i-10:i+11,j-10:j+11,0]
        H1 = a[i-10:i+11,j-10:j+11,1]
        H2 = a[i-10:i+11,j-10:j+11,2]
        img_filtered[i,j,0] = np.mean(H0)
        img_filtered[i,j,1] = np.mean(H1)
        img_filtered[i,j,2] = np.mean(H2)
     
```

```python
cv2.imshow('image',img_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()  
```

```python
cv2.imwrite('output.jpg',img_filtered)
cv2.destroyAllWindows()
```
